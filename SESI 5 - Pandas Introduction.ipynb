## Combining Multiple Datasets

```python
further_sity_data = pd.DataFrame(
    {'revenue': [700, 3400], 'employee_count': [2,2]},
    index = ['New York', 'Barcelona']
)
```

```python
further_sity_data
```

```python
city_data
```

```python
all_city_data = pd.concat([city_data, further_sity_data], sort=False)

all_city_data
```

```python
city_countries = pd.DataFrame({
    'country': ['Holland', 'Japan', 'Holland', 'Canada', 'Spain'],
    'capital': [1,1,0,0,0]},
    index= ['Amsterdam', 'Tokyo', 'Rottererdam', 'Toronto', 'Barcelona']
)

city_countries
```

```python
cities = pd.concat([all_city_data, city_countries], sort=False, axis=1)

cities
```

```python
pd.concat([all_city_data, city_countries], sort=False, axis=1, join='inner')
```

```python
countries = pd.DataFrame({
    'population_millions' : [17, 127, 37],
    'continent' : ['Europe', 'Asia', 'North America']
}, index = ['Holland', 'Japan', 'Canada'])
```

```python
countries
```

```python
cities
```

```python
pd.merge(cities, countries, left_on = 'country', right_index = True)
```

.merge() melakukan inner join secara default. jika ingin memasukkan semua kota dalam hasil, maka perlu memberikan paramter how:

```python
pd.merge(
    cities,
    countries,
    left_on = 'country',
    right_index = True,
    how = 'left'
)
```
